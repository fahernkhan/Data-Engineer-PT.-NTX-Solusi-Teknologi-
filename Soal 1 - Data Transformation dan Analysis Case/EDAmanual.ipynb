```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Import dataset
df = pd.read_csv('ecommerce-session-bigquery.csv')

# Menampilkan lima baris pertama dataset
df.head()
```

```python
# Informasi dasar tentang dataset
df.info()

# Statistik deskriptif
df.describe()
```

```python
# Menyiapkan kolom 'totalTransactionRevenue' dengan nilai 0 jika NaN
df['totalTransactionRevenue'] = df['totalTransactionRevenue'].fillna(0)

# Mengelompokkan data berdasarkan produk dan menghitung pendapatan total untuk setiap produk
top_products = df.groupby('v2ProductName')['totalTransactionRevenue'].sum().sort_values(ascending=False).head(10)

# Menampilkan 10 produk teratas
top_products
```

```python
from scipy import stats

# Menghitung jumlah transaksi untuk setiap produk
transaction_counts = df.groupby('v2ProductName')['transactions'].sum()

# Menghitung z-score untuk mendeteksi anomali
z_scores = stats.zscore(transaction_counts)

# Menemukan anomali (misalnya, z-score > 3 atau z-score < -3)
anomalies = transaction_counts[(z_scores > 3) | (z_scores < -3)]

# Menampilkan produk-produk dengan anomali transaksi
anomalies
```

```python
from scipy import stats

# Menghitung jumlah transaksi untuk setiap produk
transaction_counts = df.groupby('v2ProductName')['transactions'].sum()

# Menghitung z-score untuk mendeteksi anomali
z_scores = stats.zscore(transaction_counts)

# Menemukan anomali (misalnya, z-score > 3 atau z-score < -3)
anomalies = transaction_counts[(z_scores > 3) | (z_scores < -3)]

# Menampilkan produk-produk dengan anomali transaksi
anomalies
```

```python
# Mengelompokkan data berdasarkan kota dan menghitung pendapatan total untuk setiap kota
top_cities = df.groupby('city')['totalTransactionRevenue'].sum().sort_values(ascending=False).head(10)

# Mengelompokkan data berdasarkan negara dan menghitung pendapatan total untuk setiap negara
top_countries = df.groupby('country')['totalTransactionRevenue'].sum().sort_values(ascending=False).head(10)

# Menampilkan 10 kota dan negara teratas
print("Top 10 Cities:")
top_cities
```

```python
print("Top 10 Countries:")
top_countries
```

```python
# Visualisasi produk teratas berdasarkan total pendapatan
plt.figure(figsize=(10,6))
sns.barplot(x=top_products.values, y=top_products.index)
plt.title('Top 10 Products by Total Transaction Revenue')
plt.xlabel('Total Transaction Revenue')
plt.ylabel('Product')
plt.show()

# Visualisasi kota teratas berdasarkan total pendapatan
plt.figure(figsize=(10,6))
sns.barplot(x=top_cities.values, y=top_cities.index)
plt.title('Top 10 Cities by Total Transaction Revenue')
plt.xlabel('Total Transaction Revenue')
plt.ylabel('City')
plt.show()
```
