```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Memuat dataset
df = pd.read_csv('ecommerce-session-bigquery.csv')

# Menampilkan 5 baris pertama data
df.head()
```

```python
# Menampilkan informasi dasar tentang dataset
df.info()
```

```python
# Menampilkan statistik deskriptif
df.describe()
```

```python
# Cek kolom yang masih memiliki nilai hilang
missing_data = df.isnull().sum()
missing_data[missing_data > 0]
```

```python
# Mengisi nilai kosong dengan 0 pada kolom numerik seperti totalTransactionRevenue
df['totalTransactionRevenue'].fillna(0, inplace=True)
df['transactions'].fillna(0, inplace=True)
df['timeOnSite'].fillna(0, inplace=True)
df['pageviews'].fillna(0, inplace=True)

# Mengecek kembali apakah data masih ada yang hilang
df.isnull().sum()
```

```python
# Mengisi nilai hilang pada kolom numerik dengan rata-rata
df['sessionQualityDim'].fillna(df['sessionQualityDim'].mean(), inplace=True)
df['productQuantity'].fillna(df['productQuantity'].mean(), inplace=True)

# Atau, mengisi dengan median
df['productQuantity'].fillna(df['productQuantity'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
# Mengisi nilai hilang pada kolom kategorikal dengan 'Unknown'
df['currencyCode'].fillna('Unknown', inplace=True)
df['pageTitle'].fillna('Unknown', inplace=True)

# Mengisi nilai hilang dengan modus (nilai yang paling sering muncul)
df['currencyCode'].fillna(df['currencyCode'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
# Mengubah kolom date menjadi tipe datetime
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')

# Memeriksa hasil
df[['date']].head()
```

## Analysis

```python
# Menyimpan revenue per country
revenue_per_country = df.groupby('country')['totalTransactionRevenue'].sum().sort_values(ascending=False)

# Menampilkan 10 negara teratas berdasarkan revenue
revenue_per_country.head(10)
```

```python
# Menghitung jumlah transaksi per channel grouping
transactions_per_channel = df.groupby('channelGrouping')['transactions'].sum()

# Menampilkan hasil
transactions_per_channel
```

```python
# Plotting Revenue per Country
plt.figure(figsize=(10, 6))
sns.barplot(x=revenue_per_country.index[:10], y=revenue_per_country.values[:10])
plt.title('Top 10 Countries by Total Transaction Revenue')
plt.xlabel('Country')
plt.ylabel('Total Transaction Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting Transactions per Channel Grouping
plt.figure(figsize=(10, 6))
sns.barplot(x=transactions_per_channel.index, y=transactions_per_channel.values)
plt.title('Total Transactions per Channel Grouping')
plt.xlabel('Channel Grouping')
plt.ylabel('Total Transactions')
plt.xticks(rotation=45)
plt.show()
```

```python
# Menghitung rata-rata time on site per country
avg_time_on_site = df.groupby('country')['timeOnSite'].mean().sort_values(ascending=False)

# Menampilkan hasil
avg_time_on_site.head(10)
```

```python
# Menghitung pageviews per session quality
pageviews_per_quality = df.groupby('sessionQualityDim')['pageviews'].mean()

# Plotting hasil
plt.figure(figsize=(10, 6))
sns.barplot(x=pageviews_per_quality.index, y=pageviews_per_quality.values)
plt.title('Average Pageviews per Session Quality')
plt.xlabel('Session Quality')
plt.ylabel('Average Pageviews')
plt.show()
```

# Kesiumpulan


## #1. Mengidentifikasi Produk Teratas Berdasarkan Total Transaction Revenue per Hari

```python
# Menghitung total revenue per product per hari
product_revenue_per_day = df.groupby(['v2ProductName', 'date'])['totalTransactionRevenue'].sum().reset_index()

# Mengidentifikasi produk dengan pendapatan tertinggi
top_products_per_day = product_revenue_per_day.groupby('v2ProductName')['totalTransactionRevenue'].sum().nlargest(10)

# Visualisasi Top 10 Produk Berdasarkan Revenue
plt.figure(figsize=(10, 6))
top_products_per_day.plot(kind='bar', color='skyblue')
plt.title('Top 10 Products by Total Transaction Revenue')
plt.xlabel('Product Name')
plt.ylabel('Total Transaction Revenue')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## #2. Mendeteksi Anomali dalam Jumlah Transaksi per Produk

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Misalkan data transaksi sudah dikelompokkan per hari
# Contoh dataset (asumsi)
# df berisi kolom 'date' dan 'transactions'

# Menghitung jumlah transaksi per hari
daily_transactions = df.groupby('date')['transactions'].sum().reset_index()

# Hitung rata-rata dan standar deviasi
mean_transactions = daily_transactions['transactions'].mean()
std_transactions = daily_transactions['transactions'].std()

# Hitung z-score untuk setiap transaksi
daily_transactions['z_score'] = (daily_transactions['transactions'] - mean_transactions) / std_transactions

# Tentukan threshold untuk anomali (misalnya z-score lebih dari 3 atau kurang dari -3 dianggap anomali)
threshold = 3
daily_transactions['anomaly'] = daily_transactions['z_score'].apply(lambda x: 1 if x > threshold or x < -threshold else 0)

# Pisahkan data yang anomali
anomalies = daily_transactions[daily_transactions['anomaly'] == 1]

# Visualisasi hasil
plt.figure(figsize=(12, 6))
plt.plot(daily_transactions['date'], daily_transactions['transactions'], label='Transactions')
plt.scatter(anomalies['date'], anomalies['transactions'], color='red', label='Anomalies', zorder=5)
plt.title('Transactions Over Time with Detected Anomalies (Z-Score)')
plt.xlabel('Date')
plt.ylabel('Number of Transactions')
plt.legend()
plt.show()
```

### Mengidentifikasi Kota atau Provinsi yang Paling Menguntungkan Berdasarkan Total Transaction Revenue

```python
# Menghitung total transaction revenue per city
revenue_per_city = df.groupby('city')['totalTransactionRevenue'].sum().reset_index()

# Menyortir kota berdasarkan revenue tertinggi
top_cities = revenue_per_city.nlargest(10, 'totalTransactionRevenue')

# Visualisasi Kota Teratas Berdasarkan Revenue
plt.figure(figsize=(10, 6))
sns.barplot(x='totalTransactionRevenue', y='city', data=top_cities, palette='Blues_d')
plt.title('Top 10 Cities by Total Transaction Revenue')
plt.xlabel('Total Transaction Revenue')
plt.ylabel('City')
plt.tight_layout()
plt.show()
```
